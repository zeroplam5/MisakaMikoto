언어 : 인공지능 번역서비스 '플리토' 언어 데이터를 이용해 자동번역 서비스를 제공하는 기계

음성 : 위의 플리토는 음성언어나 말의 어투, 목소리의 톤을 통해 문맥을 파악하는 능력 역시 가지고 있다. 
이 외에도 여러 목소리 데이터를 저장해 인공적인 목소리를 만들어내는 시스템도 있다. 네이버 클로바나 인터넷 방송 스트리밍 사이트의 목소리(트위치 후원 목소리)가 대표적인 예이다.

이미지: 현재 인공지능이 가장 많은 개척을 이루어낸 분야이다. vivino 라는 어플은 와인 라벨의 이미지들을 저장하여 학습한 후 고객들에게 와인라벨을 찍기만 하면 와인에 대한 후기나 정보 등을 알려주는 기능을 한다.

자율주행: 현재 자율주행시스템은 대부분 시험운행 단계에 있다. 국내에서는 세종시에서 자율주행버스의 운행이 시작되어 조치원 세종테크노파크 등에서 실제로 운영되고 있다.
